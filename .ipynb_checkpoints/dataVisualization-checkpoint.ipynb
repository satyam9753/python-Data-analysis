 Few common background options include: whitegrid (as shown above), dark (for solid grey color), white (default) and ticks.

```python
sns.set_context("paper")
sns.set_style("dark")
sns.boxplot(x="day", y="total_bill", data=tips)
sns.despine(bottom = True)
```

```python
import numpy as np
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    
    for i in range(1,7):
        plt.plot(x, np.sin(x + i*0.5)*(7-i)*flip)
```

```python
sns.set(style="whitegrid", rc={"grid.linewidth":1.0})
sns.set_context("poster", font_scale=0.75, rc={"lines.linewidth":5.0})
sinplot()
```
